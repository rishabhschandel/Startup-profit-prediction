```python
## Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## Importing Dataset
df = pd.read_csv("50_Startups.csv")
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan,strategy="mean")
X[:,:3] = si.fit_transform(X[:,:3])
X
```

```python
##Encoding categorical column

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop='first'), [3])], remainder='passthrough')
X = ct.fit_transform(X)
X
```

```python
##Train test split

from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2)
```

```python
print(train_X)
```

```python
print(test_X)
```

```python
print(train_y)
```

```python
print(test_y)
```

```python
##Training on Train set
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(train_X, train_y)
```

```python
#Testing on test set

y_pred = regressor.predict(test_X)
print(np.concatenate((y_pred.reshape(-1,1),test_y.reshape(-1,1)), axis=-1))
```

```python
##Measuring the performance
from sklearn.metrics import r2_score
print(r2_score(test_y, y_pred))
```

```python
d1 = y_pred - test_y
d2 = y_pred - test_y.mean()
r_squared = 1 - (d1.dot(d1)/d2.dot(d2))
r_squared
```

```python
## Using k-fold cross validation
from sklearn.model_selection import cross_val_score

scores = cross_val_score(regressor, train_X, train_y, scoring="r2", cv=5)
print(scores)
print(scores.mean())
print(scores.std())
```
